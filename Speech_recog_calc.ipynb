```python
import speech_recognition as sr
import pyttsx3
import re

def speak(text):
    """Convert text to speech."""
    engine = pyttsx3.init()
    engine.say(text)
    engine.runAndWait()

def recognize_speech():
    """Capture speech input and convert it to text."""
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        print("Speak now...")
        recognizer.adjust_for_ambient_noise(source)
        try:
            audio = recognizer.listen(source, timeout=5)
            query = recognizer.recognize_google(audio).lower()
            print("You said:", query)
            return query
        except sr.UnknownValueError:
            speak("Sorry, I couldn't understand.")
        except sr.RequestError:
            speak("There was an issue with the speech recognition service.")
        except Exception as e:
            speak("An error occurred.")
            print(e)
    return None

def extract_numbers_and_operation(query):
    """Extract numbers and operation from the spoken query."""
    pattern = r'(\d+)\s*([\+\-\/]|plus|minus|into|multiplied by|times|divided by|by)\s(\d+)'
    match = re.search(pattern, query)

    if match:
        num1, operation, num2 = match.groups()
        num1, num2 = int(num1), int(num2)
        return num1, operation, num2
    return None

def calculate(num1, operation, num2):
    """Perform the mathematical operation."""
    if operation in ["plus", "+", "add"]:
        return num1 + num2
    elif operation in ["minus", "-", "subtract"]:
        return num1 - num2
    elif operation in ["into", "*", "multiplied by", "times"]:
        return num1 * num2
    elif operation in ["divided by", "/", "by"]:
        return num1 / num2 if num2 != 0 else "undefined"
    else:
        return None

def main():
    query = recognize_speech()
    if query:
        extracted = extract_numbers_and_operation(query)
        if extracted:
            num1, operation, num2 = extracted
            result = calculate(num1, operation, num2)
            if result is not None:
                answer = f"The answer is {result}."
                print(answer)
                speak(answer)
            else:
                speak("Sorry, I couldn't perform the calculation.")
        else:
            speak("Sorry, I couldn't recognize the operation.")

if __name__ == "__main__":
    main()
```
